## P - E

```python
pe_nh_ghg_cube = iris.load_cube(pe_nh_ghg_file)
pe_sh_ghg_cube = iris.load_cube(pe_sh_ghg_file)
pe_global_ghg_cube = iris.load_cube(pe_global_ghg_file)
```

```python
pe_nh_ghg_cube
```

```python
pe_nh_ghg_cube_smooth = pe_nh_ghg_cube.rolling_window('time', iris.analysis.MEAN, 10)
pe_sh_ghg_cube_smooth = pe_sh_ghg_cube.rolling_window('time', iris.analysis.MEAN, 10)
pe_global_ghg_cube_smooth = pe_global_ghg_cube.rolling_window('time', iris.analysis.MEAN, 10)
```

```python
qplt.plot(pe_nh_ghg_cube_smooth.coord('time'), pe_nh_ghg_cube_smooth - pe_nh_ghg_cube_smooth.data.mean(), label='NH, historicalGHG', color='red', linestyle='solid')
qplt.plot(pe_sh_ghg_cube_smooth.coord('time'), pe_sh_ghg_cube_smooth - pe_sh_ghg_cube_smooth.data.mean(), label='SH, historicalGHG', color='blue', linestyle='solid')
qplt.plot(pe_global_ghg_cube_smooth.coord('time'), pe_global_ghg_cube_smooth - pe_global_ghg_cube_smooth.data.mean(), label='Global, historicalGHG', color='green', linestyle='solid')
plt.legend()
plt.show()
```

```python
pe_nh_aa_cube = iris.load_cube(pe_nh_aa_file)
pe_sh_aa_cube = iris.load_cube(pe_sh_aa_file)
pe_global_aa_cube = iris.load_cube(pe_global_aa_file)
```

```python
pe_nh_aa_cube_smooth = pe_nh_aa_cube.rolling_window('time', iris.analysis.MEAN, 10)
pe_sh_aa_cube_smooth = pe_sh_aa_cube.rolling_window('time', iris.analysis.MEAN, 10)
pe_global_aa_cube_smooth = pe_global_aa_cube.rolling_window('time', iris.analysis.MEAN, 10)
```
