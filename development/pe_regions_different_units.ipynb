```python
import iris
import matplotlib.pyplot as plt
```

```python
ghg_file = '/g/data/r87/dbi599/CMIP6/DAMIP/CSIRO/ACCESS-ESM1-5/hist-GHG/r1i1p1f1/Ayr/pe/gn/v20200615/pe-region-sum_Ayr_ACCESS-ESM1-5_hist-GHG_r1i1p1f1_gn_185001-202012-cumsum.nc'
control_file = '/g/data/r87/dbi599/CMIP6/CMIP/CSIRO/ACCESS-ESM1-5/piControl/r1i1p1f1/Ayr/pe/gn/v20191214/pe-region-sum_Ayr_ACCESS-ESM1-5_piControl_r1i1p1f1_gn_010101-100012-cumsum.nc'

ghg_cube = iris.load_cube(ghg_file, 'precipitation minus evaporation flux')
control_cube = iris.load_cube(control_file, 'precipitation minus evaporation flux')

regions = ['SH precip', 'SH evap', 'tropical precip', 'NH evap', 'NH precip']

for index, region in enumerate(regions):
    plt.plot(ghg_cube.data[:, index, -1], color='red')
    plt.plot(control_cube.data[0:171, index, -1], color='grey')
    plt.show()

    ghg_end = ghg_cube.data[-1, index, -1]
    control_end = control_cube.data[170, index, -1]

    print(region, ((ghg_end - control_end) / control_end) * 100, '%')
```

Over the course of the ACCESS-ESM1-5 GHG-only experiment, the global PE region cumulative totals are 1-3% larger than for the piControl experiment.
