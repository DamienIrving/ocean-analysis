```python
import iris
import iris.plot as iplt
import matplotlib.pyplot as plt
```

```python
wfo_file = '/g/data/oi10/replicas/CMIP6/CMIP/CAS/FGOALS-f3-L/piControl/r1i1p1f1/Omon/wfo/gn/v20191028/wfo_Omon_FGOALS-f3-L_piControl_r1i1p1f1_gn_100001-109912.nc'
wfo_cube = iris.load_cube(wfo_file)
```

```python
wfo_cube
```

```python
plot_cube = wfo_cube.collapsed('time', iris.analysis.MEAN)
```

```python
plot_cube
```

```python
vmin = -0.0003
vmax = 0.0003
#iplt.pcolormesh(plot_cube, cmap='RdBu_r', vmin=vmin, vmax=vmax)
plt.pcolormesh(plot_cube.data, cmap='RdBu_r', vmin=vmin, vmax=vmax)
cb = plt.colorbar()
cb.set_label(str(plot_cube.units))
    
plt.show()
```
