```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
% matplotlib inline
```

```python
# latent / sensible / ocean

data_points = {'CCSM4': (35, 25, 40),
               'CSIRO-Mk3-6-0': (60, 20, 20),
               'FGOALS-g2': (60, 30, 10),
               'GFDL-CM3': (50, 25, 25),
               'GFDL-ESM2M': (70, 20, 10),
               'GISS-E2-H, p1': (33.3, 33.3, 33.3),
               'GISS-E2-H, p3': (33, 33, 33),
               'GISS-E2-R, p1': (40, 40, 20), 
               'GISS-E2-R, p3': (50, 25, 25),
               'IPSL-CM5A-LR': (50, 45, 5),
               'NorESM1-M': (20, 35, 45)
              }
```

```python
fig = plt.figure(figsize=(11,8))
ax = fig.add_subplot(111, projection='3d')

for model_name, data in data_points.items():
    x, y, z = data
    ax.plot([x], [y], [z], label=model_name, marker='o')

ax.set_xlabel('latent heat flux (%)')
ax.set_ylabel('sensible heat flux (%)')
ax.set_zlabel('heat flux into ocean (%)')

ax.legend(loc=3)
plt.title('Estimated compensation for skewed net radiative flux')
plt.show()
```
