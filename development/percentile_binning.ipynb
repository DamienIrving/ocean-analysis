```python
import iris
import matplotlib.pyplot as plt
import numpy as np
```

```python
pct_file = '/g/data/r87/dbi599/CMIP6/CMIP/CSIRO-ARCCSS/ACCESS-CM2/historical/r1i1p1f1/Omon/water-mass/gn/v20191108/water-mass-pct_Omon_ACCESS-CM2_historical_r1i1p1f1_gn_185001-185912.nc'
```

```python
pct_tbin_cube = iris.load_cube(pct_file, 'Ocean Grid-Cell Volume binned by temperature')
pct_sbin_cube = iris.load_cube(pct_file, 'Ocean Grid-Cell Volume binned by salinity')
tcube = iris.load_cube(pct_file, 'sea_water_potential_temperature')
scube = iris.load_cube(pct_file, 'sea_water_salinity')
```

```python
ax1 = plt.subplot()

percentiles = pct_tbin_cube.coord('Sea Water Potential Temperature percentile bins').points
ax1.plot(pct_tbin_cube[1, :, -1].data, percentiles)
ax1.set_ylabel('temperature percentile (%)')
ax1.set_xlabel('volume ($m^3$)')

plt.show()
```

```python
ax1 = plt.subplot()

percentiles = tcube.coord('Sea Water Potential Temperature percentile').points
ax1.plot(tcube[1, :].data, percentiles, marker='o')
ax1.set_ylabel('temperature percentile (%)')
ax1.set_xlabel('temperature (C)')

plt.show()
```

```python
ax1 = plt.subplot()

percentiles = pct_sbin_cube.coord('Sea Water Salinity percentile bins').points
ax1.plot(pct_sbin_cube[1, :, -1].data, percentiles)
ax1.set_ylabel('salinity percentile (%)')
ax1.set_xlabel('volume ($m^3$)')

plt.show()
```

```python
ax1 = plt.subplot()

percentiles = scube.coord('Sea Water Salinity percentile').points
ax1.plot(scube[1, :].data, percentiles, marker='o')
ax1.set_ylabel('salinity percentile (%)')
ax1.set_xlabel('salinity (C)')

plt.show()
```
